## Imports

```python
import pandas as pd
```

```python
df = pd.read_csv("harvard_repository_clean.csv")
df
```

## Value Counts

```python
authors = df["clean_author"].value_counts()
authors
```

```python
places = df["clean_place"].value_counts()
places
```

```python
publishers = df["clean_publisher"].value_counts()
publishers
```

```python
dates = df["clean_dates"].value_counts()
dates
```

```python
topics = df.topic.value_counts()
topics
```

```python
repos = df.repository.value_counts()
repos
```

## Dates

```python
df_dates = pd.DataFrame(dates)
df_dates
```

```python
df_dates.to_csv('harvard_dates.csv')
```

```python
decades = []
for year in df["clean_dates"].astype(str):
    decade = ((year[:-1] + "0"))
    decades.append(decade)
decades
df["decades"] = decades
df.head()
```

```python
df["decades_date"] = pd.to_datetime(df['decades'], format='%Y')
df["decades_date"]
```

```python
decade = df["decades_date"].value_counts()
decade
```

```python
df_decades = pd.DataFrame(decade)
df_decades
df_decades.to_csv('harvard_decades.csv')
```

```python
df["clean_dates_new"] = pd.to_datetime(df['clean_dates'], format='%Y')
df["clean_dates_new"]
```

```python
years = df["clean_dates_new"].value_counts()
years
```

```python
df_years = pd.DataFrame(years)
df_years
```

```python
df_years.to_csv('harvard_years.csv')
```
