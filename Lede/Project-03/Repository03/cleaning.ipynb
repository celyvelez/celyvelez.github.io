## Imports

```python
import pandas as pd
```

```python
df = pd.read_csv("harvard_repository.csv")
df
```

## Cleaning

```python
clean_author = []
for author in df.author:
    if len((author.strip(".").replace(" (Venancio González)","").replace(" (Medellin, Colombia)","").replace(", Colombia)","").split(", "))) == 1:
        name = ((author.strip(".").replace(" (Venancio González)","").replace(" (Medellin, Colombia)","").replace(", Colombia)","").split(", "))[0])
    elif len((author.strip(".").replace(" (Venancio González)","").replace(" (Medellin, Colombia)","").replace(", Colombia)","").split(", "))) == 2:
        firstname = (author.strip(".").replace(" (Venancio González)","").replace(" (Medellin, Colombia)","").replace(", Colombia)","").split(", "))[1]
        lastname = (author.strip(".").replace(" (Venancio González)","").replace(" (Medellin, Colombia)","").replace(", Colombia)","").split(", "))[0]
        name = firstname + " " + lastname
    clean_author.append(name)
# clean_author
df["clean_author"] = clean_author
df.head()
```

```python
clean_place = []
for place in df.place:
    new_place = ((place.strip().replace(" :","").replace("[","").replace("]","").replace("Mandada reimprimie en ","").replace(" Imprenta de Echeverria Hermanos","").replace("Cartajena","Cartagena").replace("Bogota","Bogotá").replace("Bogata","Bogotá").replace("Santafé de ","").replace(", Colombia","").replace("Bogotá́","Bogotá").replace(" República de Colombia","").replace(",","").replace("Bogotà","Bogotá").replace("?","").replace("Medellin","Medellín").replace(" de Indias","").replace("Santa Fé de ","").replace(" Colombia","").replace("Santafé de ","")))
    clean_place.append(new_place)
# clean_place
df["clean_place"] = clean_place
df.head()
```

```python
clean_publisher = []
for publisher in df.publisher:
    new_publisher = (publisher.strip(",").replace("[s.n.]","No data").replace("s.n.","No data").replace("[publisher not identified]","No data").replace("Imp. de Echeverría hnos.","Imprenta de Echeverría Hermanos").replace("Reimpresa por Echeverria Hermanos","Imprenta de Echeverría Hermanos").replace("Impr.","Imprenta").replace("Imp.","Imprenta").replace("Ed.","Editorial").replace("En la impr. del Gobierno constitucional, por Calvo","Imprenta Nacional").replace("Por el cuerpo de oficiales del 6. Batallon Ligero, imp. de Jose Casanova","Imprenta de Jose Casanova").replace("En la imprenta de D. Bruno Espinosa, por D. Nicomedes Lora","Imprenta de D. Bruno Espinosa").replace("Editorial Oveja Negra","Oveja Negra").replace("Editorial La Oveja Negra","Oveja Negra").replace("Oveja Negra","Editorial La Oveja Negra").replace("Imprenta nacional","Imprenta Nacional").replace("Imprenta del Estado","Imprenta Nacional").replace("Impreso por","Imprenta de").replace("reimpreso por ","").replace("Bog. ","").replace("(","").replace(")","").replace("Grupo Editorial Norma","Editorial Norma"))
    clean_publisher.append(new_publisher)
# clean_publisher
df["clean_publisher"] = clean_publisher
df.head()
```

```python
clean_dates = []
for dt in df.date:
    new_dates = (dt.replace(" - 9999","").replace(" - uuuu","").replace(" - 1965","").replace("año de ","").replace("-18","").replace("?","").replace(" - 1uuu","").replace(" - 1936","").replace(" - 19uu","").replace(" - 1828","").replace(" - 1869","").replace(" - 1924","").replace(" - 1955","").replace("196u","1960").replace("19uu","1900"))
    clean_dates.append(new_dates)
# clean_dates
df["clean_dates"] = clean_dates
df.head()
```

```python
df.to_csv('harvard_repository_clean.csv', index=False)
```
