### List of bills

First, we get the information from the main page of the bills. 

```python
list_link = []
list_name = []
list_legis = []
list_ini = []
list_type = []
list_authors = []

df = pd.DataFrame(columns=['link','name','legislature','iniciative','type','principal','secondary','status date','status','notes','authors','abstract'])

url_base = "https://congresovisible.uniandes.edu.co/proyectos-de-ley/"

for i in range(1,14):
    # url = url_base + "?rows=100&search=&cuatrienio=8&page=" + i + "&corporacion=2&"
    url = f"{url_base}?rows=100&search=&cuatrienio=8&page={i}&corporacion=2&"
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html)
    projects = soup.select(".listadoItem")
    for el in projects:
        list_link.append(el.select(".itemHeader")[0].select("h2")[0].select("a")[0]["href"])
        list_name.append(el.select(".itemHeader")[0].text)
        list_legis.append(el.select(".itemBody")[0].select(".itemSection")[0].select("ul")[0].select("li")[1].text.replace('Legislatura:    Legislatura ', ''))
        list_ini.append(el.select(".itemBody")[0].select(".itemSection")[0].select("ul")[0].select("li")[2].text.replace('Iniciativa:    ', ''))
        list_type.append(el.select(".itemBody")[0].select(".itemSection")[0].select("ul")[0].select("li")[3].text.replace('Tipo de proyecto:    ', ''))
        list_authors.append(el.select(".itemBody")[0].select(".itemSection")[1].select("p")[1].text)
        

df['link'] = list_link
df['name'] = list_name
df['legislature'] = list_legis
df['iniciative'] = list_ini
df['type'] = list_type
df['authors'] = list_authors


df
```

```python
list_abstract = []

# for link in df_link.head(300):
# for link in df_link.head(600).tail(300):
# for link in df_link.tail(639).head(300):
for link in df_link.tail(339):
    url_link = url_base + link
    response_link = requests.get(url_link)
    html_link = response_link.text
    soup_link = BeautifulSoup(html_link)
    abstract = soup_link.select(".TextoFormateado")

    print(abstract[0].text)
    list_abstract.append(abstract[0].text)

    sleep(1)
```
