### csv to Data Frame

```python
df_old = pd.read_csv("proyectos_ley.csv")
df = df_old.drop(columns=['Unnamed: 0'])
```

```python
df['combined'] = df.name + ' ' + df.abstract
```

### List of verbs 
In the combined column.

```python
verb_list = []

for tx in df.combined:
    doc = nlp(tx)
    for w in doc:
        if w.pos_ == 'VERB':
            verb_list.append(w)

for v in verb_list:
    print(v)
```

Note: the duplicates were removed on Google Sheets and the new list was added to the stopwords-es.txt file.
